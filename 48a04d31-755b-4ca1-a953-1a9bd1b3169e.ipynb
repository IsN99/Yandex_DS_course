```python
pivot_table_sales_platforms_years = filtered_platform_data.pivot_table(
    index='year_of_release', columns='platform', values='sales_total', aggfunc='sum')

display(pivot_table_sales_platforms_years.head(5))
display(pivot_table_sales_platforms_years.tail(5))

pivot_table_sales_platforms_years.plot(figsize=(12, 6))
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –≥–æ–¥–∞–º –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º')
plt.xlabel('–ì–æ–¥ —Ä–µ–ª–∏–∑–∞')
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
pivot_platforms_years = filtered_platform_data.pivot_table(
    index='platform', values='year_of_release', aggfunc=['max','min'])

pivot_platforms_years.columns = ['max', 'min']

# –° —É—á–µ—Ç–æ–º –≥—Ä–∞–Ω–∏—Ü –¥–∏–∞–ø–∞–∑–æ–Ω–∞
pivot_platforms_years["life_expectancy"] = pivot_platforms_years['max'] - pivot_platforms_years['min'] + 1
display(pivot_platforms_years)

average_life_expectancy = pivot_platforms_years['life_expectancy'].mean()
print("–°—Ä–µ–¥–Ω—è—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∂–∏–∑–Ω–∏ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º: ", average_life_expectancy)
```

<font color='DarkMagenta'>üëâ –ö–∞–∫-—Ç–æ –Ω–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å —Å—Ä–æ–∫ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã "–Ω–∞ –≥–ª–∞–∑–æ–∫", –µ–≥–æ —Ä–∞—Å—á–µ—Ç –∏–∑ –¥–∞–Ω–Ω—ã—Ö —Ç–∞–±–ª–∏—Ü—ã –±—ã–ª –±—ã –±–æ–ª–µ–µ —É–º–µ—Å—Ç–µ–Ω, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∫–∞–∫ –ø–æ–∫–∞–∑–∞–Ω–æ –Ω–∏–∂–µ. –í–æ–ø—Ä–æ—Å –±—Ä–∞—Ç—å –ª–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞, –∏–ª–∏ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞—Ç—å –ø–µ—Ä–∏–æ–¥–æ–º –Ω–∞—á–∞–ª–∞ —Ä–æ—Å—Ç–∞ –ø—Ä–æ–¥–∞–∂, –æ—Å—Ç–∞–µ—Ç—Å—è –æ—Ç–∫—Ä—ã—Ç—ã–º, –Ω–æ –º–æ–∂–Ω–æ –ø–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å.</font>

```python
def region_platforms(reg_sale):
    table_sales_platforms = data_actual.pivot_table(
        index='platform', values=reg_sale, aggfunc='sum'
    ).sort_values(by=reg_sale, ascending=False).head(5)

    plt.figure(figsize=(12, 6))
    plt.bar(table_sales_platforms.index, table_sales_platforms[reg_sale])
    plt.title(f'–ü–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ ({reg_sale})')
    plt.xlabel('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
    plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏')
    plt.xticks(rotation=45)

    plt.show()


sales = ["na_sales", "eu_sales", "jp_sales"]

for i in sales:
    region_platforms(i)
```
