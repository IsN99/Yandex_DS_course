### Предобработка данных

```python
data_null = data.isna().sum()
data_null[data_null!=0].sort_values()
```

```python
# # Заполним Locality_name значением "Другое" т.к. количество пропусков относительно небольшое
# # Обработаем пропуски в столбцу с локацией
# data["locality_name"] = data["locality_name"].fillna(value='Другое')
# data["locality_name"].isna().sum()

# А еще лучше выкинем эти строки
data.dropna(subset=['locality_name'], inplace=True)
data["locality_name"].isna().sum()
# Проблема пофикшена
```

<div class="alert alert-success">

**✔️ Комментарий ревьюера ✔️**

```python
data['locality_name']
```

```python
data['locality_name'] = (
    data['locality_name']
    .str.replace('ё', 'е')    
).replace(
[
    'поселок городского типа', 
    'городской поселок', 
    'коттеджный поселок', 
    'поселок при железнодорожной станции'
], 'поселок', regex=True)

result = pd.Series([elem.rsplit(' ', 1)[0] for elem in data[data['locality_name'].str.find('поселок')!=-1]['locality_name']])

result.unique()
```

```python
data['locality_name'] = (
    data['locality_name']   
).replace(
[
    'деревня', 
    'садовое товарищество', 
    'садоводческое некоммерческое товарищество'
], 'село', regex=True)

data['locality_name'].unique()
data.drop_duplicates()
data.shape
```

```python
# Создадим столбец тип локации
data['locality_type'] = data['locality_name'].apply(lambda x: 'поселок' if 'поселок' in x 
    else 'село' if 'село' in x
    else 'деревня' if 'деревня' in x
    else 'снт' if 'товарищество' in x
    else 'другое' if 'Другое' in x
    else 'город')
data['locality_type']
```

```python
# Создадим столбец удаленности от центра города
data['city_nearest_type'] = data['cityCenters_nearest'].apply(lambda x: 1 if x<10000 
    else 2 if x<20000 
    else 3 if x<40000 
    else 4)

data['city_nearest_type']
```

```python
print(data.pivot_table(index='locality_type', values='floors_total', aggfunc=['count', 'median', 'mean']))
print(data.pivot_table(index='city_nearest_type', values='floors_total', aggfunc=['count', 'median', 'mean']))
print('------------------------------------------------------------------')
print(data.pivot_table(index='locality_type', values='living_area', aggfunc=['count', 'median', 'mean']))
print(data.pivot_table(index='city_nearest_type', values='living_area', aggfunc=['count', 'median', 'mean']))
print('------------------------------------------------------------------')
print(data.pivot_table(index='locality_type', values='kitchen_area', aggfunc=['count', 'median', 'mean']))
print(data.pivot_table(index='city_nearest_type', values='kitchen_area', aggfunc=['count', 'median', 'mean']))
```

```python
print(data.pivot_table(index=['locality_type', 'city_nearest_type'], values='floors_total', aggfunc=['count', 'median', 'mean']))
print(data.pivot_table(index=['locality_type', 'city_nearest_type'], values='living_area', aggfunc=['count', 'median', 'mean']))
print(data.pivot_table(index=['locality_type', 'city_nearest_type'], values='kitchen_area', aggfunc=['count', 'median', 'mean']))
print(data.pivot_table(index=['locality_type', 'city_nearest_type'], values='ceiling_height', aggfunc=['count', 'median', 'mean']))
```

```python
# заполним медианными значениями с учетом типа локации и удаленности от центра

data['floors_total'] = data['floors_total'].fillna(data.groupby(['locality_type', 'city_nearest_type'])['floors_total'].transform('median'))
data['living_area'] = data['living_area'].fillna(data.groupby(['locality_type', 'city_nearest_type'])['living_area'].transform('median'))
data['kitchen_area'] = data['kitchen_area'].fillna(data.groupby(['locality_type', 'city_nearest_type'])['kitchen_area'].transform('median'))
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['locality_type', 'city_nearest_type'])['ceiling_height'].transform('median'))
print(data['floors_total'].isna().sum())
print(data['living_area'].isna().sum())
print(data['kitchen_area'].isna().sum())
print(data['ceiling_height'].isna().sum())
```

<div class="alert alert-success">

**✔️ Комментарий ревьюера ✔️**
   
   Хорошее решение. Как альтернатива, мы можем найти **соотношение медианной общей площади к медианной жилой**. А затем, заполнить пропуски в столбце `living_area` **отношением общей площади к найденному соотношению**.

  Со столбцом `kitchen_area` можно поступить так же, как со столбцом `living_area`.
    Данный метод основан на сильной корреляционной связи величин со столбцом `total_area`
